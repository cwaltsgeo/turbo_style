```python
turbo_color_list = []

r_memo = 48
g_memo = 18
b_memo = 59
for color in turbo_list[1:]:
    rgb_format = color[3:].replace("(", "").replace(")", "").split(",")
    r = rgb_format[0]
    g = rgb_format[1]
    b = rgb_format[2]

    code_dict = {
        "type": "algorithmic",
        "algorithm": "esriHSVAlgorithm",
        "fromColor": [
            int(r_memo),
            int(g_memo),
            int(b_memo),
            255
        ],
        "toColor": [
            int(r),
            int(g),
            int(b),
            255
        ]
    }

    r_memo = r
    g_memo = g
    b_memo = b

    turbo_color_list.append(code_dict)
```
